# Statistiques - Virgule - MNSYNLPM

Chargement des données

```python
import json
import pandas as pd
from pathlib import Path

data_dir = Path("data")

players_data = []
teams_data = []
enemies_data = []

for json_file in sorted(data_dir.glob("*.json")):
    game_id = json_file.stem
    
    with open(json_file, encoding="utf-8") as f:
        match_data = json.load(f)
    
    game_info = match_data.get("game", {})
    game_duration = game_info.get("gameDuration")
    game_creation = game_info.get("gameCreation")
    game_start = game_info.get("gameStartTimestamp")
    game_end = game_info.get("gameEndTimestamp")
    
    for player_key in match_data.keys():
        if player_key not in ["game", "virgule", "enemy"]:
            player_info = match_data[player_key]
            
            player_row = {
                "game_id": game_id,
                "gameDuration": game_duration,
                "gameCreation": game_creation,
                "gameStartTimestamp": game_start,
                "gameEndTimestamp": game_end,
                "player_name": player_key,
                "position": player_info.get("Position"),
                "championName": player_info.get("Champion", {}).get("championName"),
                "championExp": player_info.get("Champion", {}).get("championExp"),
                "championLevel": player_info.get("Champion", {}).get("championLevel"),
                "gold": player_info.get("Champion", {}).get("gold"),
                "creeps": player_info.get("Champion", {}).get("creeps"),
                "turretTakedowns": player_info.get("Champion", {}).get("turretTakedowns"),
                "itemsPurchased": player_info.get("Champion", {}).get("itemsPurchased"),
                "totalTimeSpentDead": player_info.get("Champion", {}).get("totalTimeSpentDead"),
                "longestTimeSpentLiving": player_info.get("Champion", {}).get("longestTimeSpentLiving"),
                "wardsPlaced": player_info.get("Champion", {}).get("wardsPlaced"),
                "wardsKilled": player_info.get("Champion", {}).get("wardsKilled"),
                "visionWardsPlaced": player_info.get("Champion", {}).get("visionWardsPlaced"),
                "visionScore": player_info.get("Champion", {}).get("visionScore"),

                "totalDamageDealt": player_info.get("Damage", {}).get("totalDamageDealt"),
                "physicalDamageDealt": player_info.get("Damage", {}).get("physicalDamageDealt"),
                "magicDamageDealt": player_info.get("Damage", {}).get("magicDamageDealt"),
                "trueDamageDealt": player_info.get("Damage", {}).get("trueDamageDealt"),
                "totalDamageTaken": player_info.get("Damage", {}).get("totalDamageTaken"),
                "physicalDamageTaken": player_info.get("Damage", {}).get("physicalDamageTaken"),
                "magicDamageTaken": player_info.get("Damage", {}).get("magicDamageTaken"),
                "trueDamageTaken": player_info.get("Damage", {}).get("trueDamageTaken"),
                "damageDealtToBuildings": player_info.get("Damage", {}).get("damageDealtToBuildings"),
                "damageDealtToEpicMonsters": player_info.get("Damage", {}).get("damageDealtToEpicMonsters"),
                "largestCriticalStrike": player_info.get("Damage", {}).get("largestCriticalStrike"),
                "objectivesStolen": player_info.get("Damage", {}).get("objectivesStolen"),
                "totalHeal": player_info.get("Damage", {}).get("totalHeal"),
                "totalHealsOnTeammates": player_info.get("Damage", {}).get("totalHealsOnTeammates"),

                "onMyWay": player_info.get("Ping", {}).get("onMyWay"),
                "danger": player_info.get("Ping", {}).get("danger"),
                "getBack": player_info.get("Ping", {}).get("getBack"),
                "enemyMissing": player_info.get("Ping", {}).get("enemyMissing"),
                "assistMe": player_info.get("Ping", {}).get("assistMe"),
                "retreat": player_info.get("Ping", {}).get("retreat"),
                "enemyVision": player_info.get("Ping", {}).get("enemyVision"),
                "hold": player_info.get("Ping", {}).get("hold"),
                "needVision": player_info.get("Ping", {}).get("needVision"),
                "push": player_info.get("Ping", {}).get("push"),
                "visionCleared": player_info.get("Ping", {}).get("visionCleared"),
                "allIn": player_info.get("Ping", {}).get("allIn"),
                "basic": player_info.get("Ping", {}).get("basic"),
                "totalPings": sum(player_info.get("Ping", {}).values()),

                "kills": player_info.get("KDA", {}).get("kills"),
                "deaths": player_info.get("KDA", {}).get("deaths"),
                "assists": player_info.get("KDA", {}).get("assists"),
                "doubleKills": player_info.get("KDA", {}).get("doubleKills"),
                "tripleKills": player_info.get("KDA", {}).get("tripleKills"),
                "quadraKills": player_info.get("KDA", {}).get("quadraKills"),
                "pentaKills": player_info.get("KDA", {}).get("pentaKills"),
                "largestMultiKill": player_info.get("KDA", {}).get("largestMultiKill"),
                "largestKillingSpree": player_info.get("KDA", {}).get("largestKillingSpree"),

                "firstBlood": player_info.get("First", {}).get("firstBlood"),
                "firstTower": player_info.get("First", {}).get("firstTower"),

                "Q_spell": player_info.get("Spell", {}).get("Q_spell"),
                "W_spell": player_info.get("Spell", {}).get("W_spell"),
                "E_spell": player_info.get("Spell", {}).get("E_spell"),
                "R_spell": player_info.get("Spell", {}).get("R_spell"),
                "summoner1Casts": player_info.get("Spell", {}).get("summoner1Casts"),
                "summoner2Casts": player_info.get("Spell", {}).get("summoner2Casts"),
            }
            players_data.append(player_row)
    
    virgule = match_data.get("virgule", {})
    team_row = {
        "team": "virgule",
        "game_id": game_id,
        "gameDuration": game_duration,
        "gameCreation": game_creation,
        "gameStartTimestamp": game_start,
        "gameEndTimestamp": game_end,
        "team_name": "virgule",
        "win": virgule.get("win"),
        "kills": virgule.get("kills"),
        "tower": virgule.get("tower"),
        "inhibitor": virgule.get("inhibitor"),
        "dragon": virgule.get("dragon"),
        "horde": virgule.get("horde"),
        "riftHerald": virgule.get("riftHerald"),
        "baron": virgule.get("baron"),
        "picks": ",".join(virgule.get("picks", [])),
        "bans": ",".join([b.get("championName", "") for b in virgule.get("bans", [])]),
    }
    teams_data.append(team_row)
    enemy = match_data.get("enemy", {})
    enemy_row = {
        "team": "enemy",
        "game_id": game_id,
        "gameDuration": game_duration,
        "gameCreation": game_creation,
        "gameStartTimestamp": game_start,
        "gameEndTimestamp": game_end,
        "team_name": "enemy",
        "win": enemy.get("win"),
        "kills": enemy.get("kills"),
        "tower": enemy.get("tower"),
        "inhibitor": enemy.get("inhibitor"),
        "dragon": enemy.get("dragon"),
        "horde": enemy.get("horde"),
        "riftHerald": enemy.get("riftHerald"),
        "baron": enemy.get("baron"),
        "picks": ",".join(enemy.get("picks", [])),
        "bans": ",".join([b.get("championName", "") for b in enemy.get("bans", [])]),
    }
    teams_data.append(enemy_row)

# Créer les DataFrames
df_players = pd.DataFrame(players_data)
df_team = pd.DataFrame(teams_data)

print(f"df_players: {len(df_players)} lignes, {len(df_players.columns)} colonnes")
print(f"df_team: {len(df_team)} lignes, {len(df_team.columns)} colonnes")

print("\n=== Aperçu df_players ===")
print(df_players[["game_id", "player_name", "position", "championName", "kills", "deaths", "assists"]].head())

print("\n=== Aperçu df_team ===")
print(df_team[["game_id", "team_name", "win", "kills", "picks"]].head())
```
